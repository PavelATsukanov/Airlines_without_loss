```python
#Истиное время в пути:
proj_4_1 = proj_4.merge(actual_dutation, left_on='flight_id', right_on='flight_id', suffixes=('_left', '_right'), indicator=True)
proj_4_1
```

```python
pr1=proj_4_1.merge(id_summ_amm, left_on='flight_id', right_on='flight_id', suffixes=('_left', '_right'))
pr1
```

```python
pr1 = pr1.astype({'lainer_Capacity':'int','People_on_a_board':'int'},errors='ignore')
pr1
```

```python
pr1['empty_seats'] = pr1['lainer_Capacity'] - pr1['People_on_a_board']
pr1
```

```python
pr1['perc_of_empty']=(pr1['empty_seats'] *100)/pr1['lainer_Capacity']
pr1
```

```python
pr1 = pr1.astype({'average_sum':'int','People_on_a_board':'int','actual_flight_time':'int','fuel_consumption_ton_per_hour':'int','fuel_cost':'int'},errors='ignore')
pr1
```

```python
pr1['fuel_consumption_per_min']=pr1['fuel_consumption_ton_per_hour']/60
pr1
```

```python
pr1['expenses'] = pr1['fuel_consumption_per_min'] * pr1['actual_flight_time']* pr1['fuel_cost']/1000
pr1.expenses = pr1.expenses.round()
pr1
```

```python
pr1['profit'] = pr1['average_sum'] - pr1['expenses']
pr1
```
